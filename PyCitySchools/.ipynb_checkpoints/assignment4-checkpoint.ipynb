```python
import pandas as pd

school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

 
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

# District Summary-start

```python
#Calculate the total number of schools
total_schools=school_data_complete["school_name"].nunique()
total_schools
```

```python
#Calculate the total number of students
total_student=school_data_complete["Student ID"].count() 
# ####shall we use unique? all we think all same name will be diff poeple?
total_student
```

```python
#Calculate the total budget
sumBudget=school_data_complete["budget"].unique().sum() 
### all data is different this the way i found to deal with high level df
sumBudget
sumBudget="$"+format(sumBudget,",.2f")
sumBudget
```

```python
#Calculate the average math score 
avgmath=round((school_data_complete["math_score"].mean()),6)
avgmath
```

```python
#Calculate the average reading score
avgreading=round((school_data_complete["reading_score"].mean()),6)
avgreading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
occurancegreaterequalto70math=school_data_complete[school_data_complete["math_score"]>=70].count()
occurancegreaterequalto70math# notice we retireve data is a series
occurancegreaterequalto70math.dtype
passedmath=occurancegreaterequalto70math.iloc[0]
mathpassingrate=passedmath/total_student
passingpercentagemath=print(f'{mathpassingrate*100}%')
rpassingpercentagemath=round(mathpassingrate*100,6)
rpassingpercentagemath# this is the well rounded number
mathpassingperc=print(f'{rpassingpercentagemath}%')
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
occurancegreaterequalto70reading=school_data_complete[school_data_complete["reading_score"]>=70].count()
occurancegreaterequalto70reading
passedreading=occurancegreaterequalto70reading.iloc[0]
readingpassingrate=passedreading/total_student
rreadingpassingrate=round(readingpassingrate*100,8)
rreadingpassingrate# this the round before percentage
readingpassingperc=print(f"{rreadingpassingrate}%")
```

```python
#Calculate the percentage of students who passed math **and** reading (% Overall Passing)
occuranceforboth=school_data_complete[(school_data_complete["reading_score"]>=70)&(school_data_complete["math_score"]>=70)
                                     ].count()
passingboth=occuranceforboth.iloc[0]
passingbothrate=passingboth/total_student
passingbothrate=round(passingbothrate*100,8)
passingbothrate # this the round before percentage
bothpassingperc=print(f"{passingbothrate}%")
bothpassingperc
```

```python
#Create a dataframe to hold the above results

rawtable=[{"Total Schools":total_schools,"Total Students":total_student,"Total Budget":sumBudget,
          "Average Math Score":avgmath,"Average Reading Score":avgreading,"% Passing Math":rpassingpercentagemath,
         "% Passing Reading":rreadingpassingrate,"% Overall Passing":passingbothrate}]
SummaryDataFrame=pd.DataFrame(rawtable)
SummaryDataFrame
# note if we use {} on each element it will break to multiple row
```
